# **Assessment Task 1: Basic Probability Calculations**

1,Start with a Coding Task:

```python
def joint_probability(p_A, p_B):
    """
    Calculate the joint probability of two independent events A and B.

    Parameters:
    p_A (float): Probability of event A.
    p_B (float): Probability of event B.

    Returns:
    float: Joint probability of A and B.
    """
    return p_A * p_B

def marginal_probability(p_A, p_B):
    """
    Calculate the marginal probability of event A or B.

    Parameters:
    p_A (float): Probability of event A.
    p_B (float): Probability of event B.

    Returns:
    float: Marginal probability of A or B.
    """
    return p_A + p_B - joint_probability(p_A, p_B)

def conditional_probability(p_A, p_B, joint_p):
    """
    Calculate the conditional probability of event A given event B.

    Parameters:
    p_A (float): Probability of event A.
    p_B (float): Probability of event B.
    joint_p (float): Joint probability of A and B.

    Returns:
    float: Conditional probability P(A|B).
    """
    if p_B == 0:
        return None
    return joint_p / p_B

```

# **Assessment Task 2: Bayesian Inference**


1. Interactive Bayesian Model

```python
def bayesian_inference(sensitivity, false_positive_rate, prior_disease):

    p_positive_given_disease = sensitivity
    p_positive_given_no_disease = false_positive_rate
    p_no_disease = 1 - prior_disease


    p_positive = (p_positive_given_disease * prior_disease +
                  p_positive_given_no_disease * p_no_disease)


    p_disease_given_positive = (p_positive_given_disease * prior_disease) / p_positive

    return p_disease_given_positive
```

```python
sensitivity = 0.9
false_positive_rate = 0.05
prior_disease = 0.01
```

```python
posterior_probability = bayesian_inference(sensitivity, false_positive_rate, prior_disease)
print(f"The probability that someone who tested positive actually has the disease is: {posterior_probability:.4f}")
```

```python
report = """
This result indicates that even with a test that has a sensitivity of 90% and a false positive rate of 5%, only about15.38% of individuals
who receive a positive test result actually have the disease. This highlights the importance of considering prior probabilities and the context
of medical testing when interpreting results.
"""
print(report)
```

# **Assessment Task 3: Decision-Making Under Uncertainty**

```python
import random

def simulate_investment_decision(probability_of_success, reward, penalty, iterations=1000):

    total_return = 0

    for _ in range(iterations):
        if random.random() < probability_of_success:
            total_return += reward  # Successful investment
        else:
            total_return += penalty  # Failed investment

    average_return = total_return / iterations
    return average_return
```

```python
probability_of_success = 0.7
reward = 1000
penalty = -500


average_return = simulate_investment_decision(probability_of_success, reward, penalty)
print(f"The average return after 1000 iterations is: ${average_return:.2f}")
```

```python
new_probability_of_success = 0.8
average_return_adjusted = simulate_investment_decision(new_probability_of_success, reward, penalty)
print(f"The average return after 1000 iterations with {new_probability_of_success*100}% success rate is: ${average_return_adjusted:.2f}")
```

```python
report = """
These distributions quantify the likelihood of various outcomes based on known probabilities, allowing decision-makers to assess risks effectively.
A higher probability of success typically leads to more favorable expected returns, while a lower probability increases the risk of losses.
By analyzing different probability distributions associated with potential outcomes, decision-makers can choose strategies that align with their risk tolerance and expected returns.
 """
print(report)
```

# **Assessment Task 4: Real-World Scenario**

```python
import numpy as np
import matplotlib.pyplot as plt


def p_rain(humidity, cloud_cover):


    base_probability = 0.1

    if humidity > 70:
        base_probability += 0.3
    elif humidity > 40:
        base_probability += 0.2

    if cloud_cover > 70:
        base_probability += 0.4
    elif cloud_cover > 40:
        base_probability += 0.2


    return min(base_probability, 1.0)

def simulate_rain_prediction(humidity_range, cloud_cover_range):
    probabilities = []

    for humidity in humidity_range:
        for cloud_cover in cloud_cover_range:
            prob = p_rain(humidity, cloud_cover)
            probabilities.append((humidity, cloud_cover, prob))

    return probabilities

humidity_range = np.linspace(0, 100, num=100)
cloud_cover_range = np.linspace(0, 100, num=100)


results = simulate_rain_prediction(humidity_range, cloud_cover_range)


humidity_values = [result[0] for result in results]
cloud_cover_values = [result[1] for result in results]
probability_values = [result[2] for result in results]


plt.figure(figsize=(10,6))
sc = plt.scatter(humidity_values, cloud_cover_values, c=probability_values, cmap='Blues', marker='o')
plt.colorbar(sc, label='Probability of Rain')
plt.title('Probability of Rain Based on Humidity and Cloud Cover')
plt.xlabel('Humidity (%)')
plt.ylabel('Cloud Cover (%)')
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.grid()
plt.show()
```

# **Assessment Task 5: Probability Distributions Visualization**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_binomial_distribution(n, p):
    """
    Plot the binomial distribution.

    Parameters:
    n (int): Number of trials.
    p (float): Probability of success (heads).
    """
    # Binomial distribution
    x = np.arange(0, n + 1)
    binom_dist = np.random.binomial(n, p, size=1000)

    plt.figure(figsize=(12, 6))
    sns.histplot(binom_dist, bins=n+1, kde=False, stat='density', color='blue', alpha=0.6)

    # Plotting the theoretical binomial distribution
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.xlabel('Number of Heads')
    plt.ylabel('Probability Density')

    # Plotting the probability mass function
    pmf = [np.math.comb(n, k) * (p**k) * ((1-p)**(n-k)) for k in x]
    plt.plot(x, pmf, 'ro-', label='Theoretical PMF', markersize=4)

    plt.legend()
    plt.grid()
    plt.show()

# Simulate 1000 coin flips with default parameters
n_trials = 1000  # Number of trials (coin flips)
p_heads = 0.5   # Probability of heads

# Plot Binomial Distribution
plot_binomial_distribution(n_trials, p_heads)

# Plot Normal Distribution Approximation
plot_normal_distribution(n_trials, p_heads)
```
